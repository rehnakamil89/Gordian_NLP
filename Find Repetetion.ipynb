```python
import numpy as np
import pandas as pd
import tensorflow as tf
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_excel('PlumbingPOC.xlsx', sheet_name='Plumbing') 
print(df)
```

```python
Y = df[['tech_desc','mat_skey']]
Y
```

```python
def split_comma(in_string):
    split_stings = [x.strip() for x in in_string.split(',')]
    return split_stings
```

```python
split_comma(df['tech_desc'][15])
```

```python
word_list = []
for i in Y['tech_desc']:
    word_list.extend(split_comma(i))    
```

```python
word_list
```

```python
word_list_set = list(set(word_list))
word_list_set
```
